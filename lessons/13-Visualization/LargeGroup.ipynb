FFF = Fudge's Food Finder!

Input City, Cuisine

Plot on a map all the rests where you can eat that cuisine in that city!



```python
import import_ipynb
import googleapi as google
import pandas as pd
import folium
from IPython.display import display
from ipywidgets import interact_manual

google_api_key = '751838d6b9978955679461c80a7df60bde4f093d'

@interact_manual(location="", cuisine="")
def draw_map(location, cuisine):
    geo = google.googleGeocode(google_api_key, location)
    locations = google.googlePlacesSearch(google_api_key, geo['lat'], geo['lng'],"restaurant", keyword=cuisine)
    locdf = pd.json_normalize(locations)
    fmap = folium.Map(location=(geo['lat'], geo['lng']), zoom_start=13 )

    for row in locdf.to_records():
        marker = folium.Marker(location=(row['geometry.location.lat'], row['geometry.location.lng']), popup=row['name'])
        fmap.add_child(marker)
      
    display(fmap)
    return 
```
