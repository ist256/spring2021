```python
# Call an API, return a Python data structure 
def geocode(location):
    query_string = {'q' : location, 'format': 'json'}
    url='https://nominatim.openstreetmap.org/search' 
    response = requests.get(url, params = query_string)
    response.raise_for_status()
    geodata = response.json()
    return geodata

places = geocode("Syracuse") # call API
places

geo_df = pd.DataFrame(places) # build dataframe from the Python data structure
geo_df[ geo_df['class'] == 'railway' ]
```

```python
for row in geo_df.to_records():
    if float(row['lat']) > 40:
        print(row['display_name'])
    
```
