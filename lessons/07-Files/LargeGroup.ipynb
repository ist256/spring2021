```python
total = 0
count = 0
with open("data/grades.txt","r") as grades_file:
    for line in grades_file.readlines():
        grade =int(line)
        total = total + grade
        count = count + 1
        
print(f"Your Average of {count} grades is {total/count}")
        
```

```python
def check_spelling(your_word):
    dict_file = 'dict.txt'
    with open(dict_file,"r") as df:
        for word in df.readlines():
            if your_word == word.strip():
                return True
            
        return False
                
text = input("Enter something to spellcheck: ")
words = text.split()
for word in words:
    if not check_spelling(word):
        print(f"Misspelled word: {word}")
```
