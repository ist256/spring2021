```python
import random
low = 1
high = 10
number = random.randint(low,high)
while True:  # starting a loop
    guess = int(input(f"Guess a number between {low} and {high}: "))
    if guess == number:
        break #stop looping
    else:
        print("Wrong!")
        
print("You guessed it!")
```
